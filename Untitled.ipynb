```python
def make_soup(url):       # Using BeautifulSoup to prepare parsing
    html = open(url.encode('utf-8'))
    return BeautifulSoup(html, 'lxml')

def tide_list (x, y):	  # removing non-elements context
	for element in x:
		element = ' '.join(re.findall(r'[\S.]+', element))
		y.append(element)
	return 
distractors_store = []
def create_distractor(ori_list):
    distractor_list = ori_list[:]
    if len(ori_list) == 1:
        distractor_list = distractors_store[-4]
        print(distractors_store[-4])
    else:
        while True:
            random.shuffle(distractor_list)
            for a, b in zip(ori_list, distractor_list):
                if a == b:
                    break
                else:
                    distractors_store.extend(distractor_list)
                    return distractor_list

#舍弃muti—— distractor function, 直接用 create-distractor去生成trainset里的distractor

def muti_distractor(ori_list):# Generate 10 distractors, one for trainning set, other 9 for testing set
    distractor_dict = {}
    for x in range (0,10):
        if len(ori_list) <= 4:
            distractor_dict['distractor_{0}'.format(x)] = distractors_store[x-16]
        else:
            distractor_dict['distractor_{0}'.format(x)] = create_distractor(ori_list)
    return distractor_dict 
```
